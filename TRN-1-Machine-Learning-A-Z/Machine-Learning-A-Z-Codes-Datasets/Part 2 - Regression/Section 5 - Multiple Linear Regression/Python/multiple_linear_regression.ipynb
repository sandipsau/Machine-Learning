# Multiple Linear Regression

## Assumption of Linear regression
- Analyze the data to confirm the dataset fits linear regression
- Anscombe's quartet
    - Anscombes quartet is a set of four datasets that share identical statistical propertiessuch as mean, variance, correlation, and linear regressionbut have very different distributions when visualized. It demonstrates that summary statistics alone are insufficient and that data visualization is essential for proper analysis.

    

- What the four datasets actually look like
    - Dataset I  Normal
        - Looks like a reasonable linear relationship
        - Linear regression makes sense
    - Dataset II  Curved
        - Strong non-linear pattern
        - Linear regression is misleading
    - Dataset III  Outlier-driven
        - Almost all points align except one extreme outlier
        - That one point drives the correlation
    - Dataset IV  Vertical cluster + outlier
        - Most points share the same x
        - One outlier creates the illusion of correlation  
- Why this matters (this is the lesson)
    - If you only look at:
        - mean
        - variance
        - correlation
        - regression coefficients          
    you can reach a completely wrong conclusions.
    - Visualization exposes
        - non-linearity
        - outliers
        - heteroscedasticity
        - data quality issues 
- Assumptions of Linear Regression
    - Linearity - Linear relationship between y and each X
    - Homoscedasticity  - Equal variance
    - Multivariate Normality - Normality of error distribution
    - Independence - Includes no auto correlation
    - Lack of Multicollinearity - Predictors are not correlated to each other
    - The Outlier check         

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

## Encoding categorical data

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression model on the Training set

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
